```python
txt = 'burgers,meatballs,eggs'
type(txt)
txt.split('/')
```

```python
def txt_split(txt):
    return txt.split(',')
```

```python
txt_split('python,session,machine,learning')
```

```python
df.products.head()
```

```python
data = df['products'].apply(txt_split)
data
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
encoded_df = te.fit_transform(data)
```

```python
encoded_df
```

```python
te.columns_
```

```python
data = pd.DataFrame(encoded_df, columns=te.columns_)
data
```

```python
# data.replace(True,1,inplace=True)
# data.replace(False,0,inplace=True)
# data

data.replace([True,False],[1,0], inplace=True)
```

```python
data
```

```python
data.shape
```

```python
scores = apriori(data,min_support = 0.01, use_colnames=True)
scores
```

```python
rules = association_rules(scores)
rules
```

```python
rules.sort_values(by = 'lift', ascending=False)
```
